## Задача 1. Анализ индикаторов качества государственного управления (The Worldwide Government Indicators, WGI)

```python
# Подключаем необходимые библиотеки
import matplotlib.pyplot as plt
%matplotlib inline
import pandas as pd
```

```python
# Создаем датафрейм по wgidataset и выводим его вершину
df_wgidataset = pd.read_excel(
    'wgidataset.xlsx', sheet_name='ControlofCorruption', header=14
)
df_wgidataset.head()
```

```python
# Сортируем данные по убыванию индекса
df_wgidataset.iloc[::-1]
```

```python
# Сортируем данные по столбцу Rank за 2022
df_wgidataset_sorted = df_wgidataset.sort_values('Rank.23')

# Отображаем данные
plt.figure(figsize=(30, 50))
plt.barh(
    df_wgidataset_sorted['Country/Territory'], df_wgidataset_sorted['Rank.23']
)
plt.show()
```

```python
# Создаем датафрейм по regions и выводим его вершину
df_regions = pd.read_excel('../data/A1_Descriptive_Analysis/regions.xlsx')
df_regions.head()
```

```python
# Объединяем датафреймы wgidataset и regions
df_merged = pd.merge(df_wgidataset, df_regions, on='Code')\
    .drop('Country/Territory', axis=1)\
    .set_index('Country')

# Выводим результат по региону Asia Pacific
df_merged.groupby('Region').get_group('AP')
```

```python
# Получаем индекс WGI
wgi = df_merged[df_merged.Region == 'AP'].filter(regex='Estimate|Country')

# Строим график индекса WGI
wgi.T.plot(
    color='grey', grid=1, figsize=(25,7),
    title='WGI за 1996-2022', marker='o', legend=False
)
```

```python
# Выводим страны с минимальным и максимальным значением индекса WGI
min_wgi = wgi['Estimate.23'].idxmin()
max_wgi = wgi['Estimate.23'].idxmax()

[min_wgi, max_wgi]
```

```python
# Выводим среднее значение индекса WGI
mean_wgi = wgi.mean()
mean_wgi
```

```python
# Строим график индекса WGI
wgi.T.plot(
    color='grey', grid=1, figsize=(25,7),
    title='WGI за 1996-2022', marker='o', legend=False
)

# Строим линии mean, max, min WGI и РФ
mean_wgi.plot(color='blue', marker='o', label='mean', legend=True)
wgi.T[max_wgi].plot(color='red', marker='o', label='max', legend=True)
wgi.T[min_wgi].plot(color='green', marker='o', label='min', legend=True)
df_merged.T['Russia']\
    .filter(regex='Estimate')\
    .plot(color='orange', marker='o', label='Russia', legend=True)
```

```python
# Создаем датафрейм столбцов Rank
df_ranks = df_merged.copy().filter(regex='Rank')

# Создаем датафрейм изменений по столбцу Rank и выводим результат
df_changes = df_ranks\
    .filter(items=['Rank', 'Rank.23'])\
    .pct_change(axis=1)['Rank.23']
df_changes
```

```python
# Создаем таблицу
table = pd.DataFrame(
    index=['mean_2022', 'max_2022', 'min_2022', 'Russia_2022'], 
    columns=['Регион', 'Страна', 'WGI 1996', 'WGI 2022', 'Изменение']
)

# Заполняем столбец Регион
table.loc['mean_2022', 'Регион'] = 'AP'
table.loc['max_2022', 'Регион'] = 'AP'
table.loc['min_2022', 'Регион'] = 'AP'
table.loc['Russia_2022', 'Регион'] = 'ECA'

# Заполняем столбец Страна
table.loc['mean_2022', 'Страна'] = ''
table.loc['max_2022', 'Страна'] = max_wgi
table.loc['min_2022', 'Страна'] = min_wgi
table.loc['Russia_2022', 'Страна'] = 'Russia'

# Заполняем столбец WGI 1996
table.loc['mean_2022', 'WGI 1996'] = df_ranks['Rank'].mean()
table.loc['max_2022', 'WGI 1996'] = df_ranks['Rank'][max_wgi]
table.loc['min_2022', 'WGI 1996'] = df_ranks['Rank'][min_wgi]
table.loc['Russia_2022', 'WGI 1996'] = df_ranks['Rank']['Russia']

# Заполняем столбец WGI 2022
table.loc['mean_2022', 'WGI 2022'] = df_ranks['Rank.23'].mean()
table.loc['max_2022', 'WGI 2022'] = df_ranks['Rank.23'][max_wgi]
table.loc['min_2022', 'WGI 2022'] = df_ranks['Rank.23'][min_wgi]
table.loc['Russia_2022', 'WGI 2022'] = df_ranks['Rank.23']['Russia']

# Заполняем столбец Изменение
table.loc['mean_2022', 'Изменение'] = df_changes.mean()
table.loc['max_2022', 'Изменение'] = df_changes[max_wgi]
table.loc['min_2022', 'Изменение'] = df_changes[min_wgi]
table.loc['Russia_2022', 'Изменение'] = df_changes['Russia']

table
```

```python
# Определяем функцию для получения данных по региону
def get_region_data(name):
    df_result = df_merged.copy()
    if name != 'ALL':
        df_result = df_result[df_merged.Region == name]
    
    return df_result\
        .filter(regex='Estimate.23|Country')\
        .rename(columns={'Estimate.23': name})

# Отображаем данные по регионам
plt.figure(figsize=(20,20))
plt.title('Диаграмма размаха WGI за 2022')
pd.concat(
    map(
        get_region_data, 
        ['AP', 'AME', 'ECA', 'MENA', 'SSA', 'WE/EU', 'ALL']
    ), sort=True
).boxplot()
```

## Задача 2. Анализ рынка акций

```python
# Подключаем все файлы из папки data/stock
import glob
glob_files = glob.glob('../data/A1_Descriptive_Analysis/stock/*.csv')

# Создаем хэш вида `имя:цена закрытия`
closes_hash = {}
for glob_file in glob_files:
    start_index = glob_file.rfind('/')
    end_index = glob_file.find('.csv')
    name = glob_file[start_index+1:end_index]
    
    df_file = pd.read_csv(glob_file, index_col='Date')
    closes_hash[name] = df_file['Close']

# Объединяем данные хэша и выводим их
df_closes = pd.concat(closes_hash, axis=1, sort=True)
df_closes
```

```python
# Отображаем корреляционную матрицу для всех акций
corr = df_closes.corr()
corr
```

```python
# Подключаем библиотеку для диаграмм
import seaborn as sns

# Отображаем корреляционную матрицу в виде диаграммы
plt.title('Корреляционная матрица')
sns.heatmap(corr, annot=False, square=True)
```

```python
# Выводим акции с максимальной положительной и отрицательной корреляцией
corr_filtered = corr.copy()\
    .T.drop(['NFLX'], axis=1)\
    .T.filter(regex='NFLX')
corr_max = corr_filtered.idxmax()
corr_min = corr_filtered.idxmin()

[corr_max[0], corr_min[0]]
```

```python
# Ищем число, наиболее близкое к 0
nflx_values = corr_filtered['NFLX'].values
about_zero = min(nflx_values, key=lambda x: abs(x))

# Определяем название фирмы
corr_none = corr_filtered['NFLX']
corr_none = corr_none[abs(corr_none.values - about_zero) < 1e-9]
corr_none
```

```python
# Отображаем диаграмму разброса относительно min
plt.title('Диаграмма разброса относительно min')
plt.scatter(df_closes['NFLX'], df_closes[corr_min])
```

```python
# Отображаем диаграмму разброса относительно max
plt.title('Диаграмма разброса относительно max')
plt.scatter(df_closes['NFLX'], df_closes[corr_max])
```

```python
# Отображаем диаграмму разброса относительно none
plt.title('Диаграмма разброса относительно none')
plt.scatter(df_closes['NFLX'], df_closes[corr_none.keys()])
```

```python
# Отображаем среднюю цену акций для каждого месяца
closes_mean = df_closes.T.mean()
closes_mean
```

```python
# Отображаем линии NFLX, max, min, none и average
df_closes['NFLX']\
    .plot(grid=1, figsize=(25,7), color='blue', marker='o', label='NFLX')
df_closes[corr_max]\
    .T.mean()\
    .plot(
        grid=1, figsize=(25,7), color='orange',
        marker='o', label='UBER - max'
    )
df_closes[corr_min]\
    .T.mean()\
    .plot(
        grid=1, figsize=(25,7), color='green',
        marker='o', label='TSLA - min'
    )
df_closes[corr_none.keys()]\
    .T.mean()\
    .plot(
        grid=1, figsize=(25,7), color='red',
        marker='o', label='TWLO - none'
    )
closes_mean\
    .plot(grid=1, figsize=(25,7), color='grey', marker='o', label='Average')

# Задаем настройки отображения
plt.title('График акций')
plt.legend()
```
